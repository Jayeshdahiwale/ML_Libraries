## scatter plot for random 1000 x and y coordinates

```python
import plotly.express as px
import numpy as np


# creating random data through randomint
# function of numpy.random
np.random.seed(42)

random_x= np.random.randint(1,101,1000)
random_y= np.random.randint(1,101,1000)

plot = px.scatter(random_x, random_y)
plot.layout
plot.show()
```

## draw line and scatter plots for random 100 x and y coordinates

```python
import plotly.graph_objects as go
random_x= np.random.randint(1,101,100)
random_y= np.random.randint(1,101,100)
fig = go.Figure(data=go.Scatter(x=random_x, y=random_y, mode = "lines"))

fig.show()
```

## scatter plot for random 500 x and y coordinates and style it

```python
np.random.seed(1)
```

```python


x = np.random.uniform(low=3, high=6, size=(500,))
y = np.random.uniform(low=3, high=6, size=(500,))

# Build figure
fig = go.Figure()

# Add scatter trace with medium sized markers
fig.add_trace(
    go.Scatter(
        mode='markers',
        x=x,
        y=y,
        marker=dict(
            color='LightSkyBlue',
            size=20,
            line=dict(
                color='MediumPurple',
                width=2
            )
        ),
        showlegend=True
    )
)
fig.update_xaxes(title='Random_X')
fig.update_yaxes(title='Random_Y')

fig.update_layout(title_text='Graph between Random X and Y', title_x=0.5)



fig.show()
```

## draw a scatter plot for a given dataset and show datalabels on hover

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv")
```

```python
df
```

```python
fig = go.Figure(data=go.Scatter(x=df["State"], y=df["Population"],mode="markers"))
fig.update_xaxes(title='States')
fig.update_yaxes(title='Population')

fig.update_layout(title_text='2014 USA state wise population', title_x=0.5)


fig.show()
```
