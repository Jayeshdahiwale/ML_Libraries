# SEABORN

## bar plot of sex against survived for a dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_style('darkgrid')
# Import Data
titanic_dataset = sns.load_dataset("titanic")

```

```python
titanic_dataset
```

```python
# Construct plot
sns.barplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```

## Point plot of sex against survived

```python
sns.pointplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```

## scatter plot of “day” against “total bill” in tips.csv dataset

# Violin Plot

## violin plot of “sex” against “total bill” in tips.csv dataset


```python
sns.violinplot(x="sex",y="total_bill",data=tips_dataset)
plt.show()
```

# Seaborn Box Plot

## box plot of life expectancy by continent for a data set given in a url

```python
import pandas as pd
gapmind_dataset =pd.read_csv("https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv")
```

```python
sns.boxplot(x="continent",y="lifeExp",data=gapmind_dataset)
plt.show()
```
