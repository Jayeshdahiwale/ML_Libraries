# PIE CHART

## importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# pie chart of the popularity of programming Languages.

```python
# Data to plot
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["Blue", "Yellow", "Green", "Red", "Purple", "Grey"]
# explode 1st slice
explode = (0.1, 0, 0, 0,0,0)
# Plot
plt.pie(popularity, explode=explode, labels=languages, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle= 120)

plt.axis('equal')
plt.show()
```

## create a pie chart with a title of the popularity of programming Languages.

```python
# Data to plot
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["Blue", "Yellow", "Green", "Red", "Purple", "Grey"]
# explode 1st slice
explode = (0.1, 0, 0, 0,0,0)
# Plot
plt.pie(popularity, explode=explode, labels=languages, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle= 120)

plt.axis('equal')
plt.title("Popularity of Programming Languages")
plt.show()
```
